```python
#https://powietrze.gios.gov.pl/pjp/content/health_informations/
# PM2.5
wartoscDobraPM2_5 = smogPM2_5['Wartosc']
countWartoscDobraPM2_5 = wartoscDobraPM2_5[(wartoscDobraPM2_5 >= 0) & (wartoscDobraPM2_5 <= 35)].count()

wartoscSredniaPM2_5 = smogPM2_5['Wartosc']
countWartoscSredniaPM2_5 = wartoscSredniaPM2_5[(wartoscSredniaPM2_5 >= 35.1) & (wartoscSredniaPM2_5 <= 75)].count()

wartoscZlaPM2_5 = smogPM2_5['Wartosc']
countWartoscZlaPM2_5 = wartoscZlaPM2_5[wartoscZlaPM2_5 >= 75.1].count()

############ 
# PM10

wartoscDobraPM10 = smogPM10['Wartosc']
countWartoscDobraPM10 = wartoscDobraPM10[(wartoscDobraPM10 >= 0) & (wartoscDobraPM10 <= 50)].count()

wartoscSredniaPM10 = smogPM10['Wartosc']
countWartoscSredniaPM10 = wartoscSredniaPM10[(wartoscSredniaPM10 >= 50.1) & (wartoscSredniaPM10 <= 110)].count()

wartoscZlaPM10 = smogPM10['Wartosc']
countWartoscZlaPM10 = wartoscZlaPM10[wartoscZlaPM10 >= 110.1].count()


#Konwersja numpy.int na regularny int
#PM2.5
valZla2_5 = countWartoscZlaPM2_5.item()
valSrednia2_5 = countWartoscSredniaPM2_5.item()
valDobra2_5 = countWartoscDobraPM2_5.item()

#PM10
valZla10 = countWartoscZlaPM10.item()
valSrednia10 = countWartoscSredniaPM10.item()
valDobra10 = countWartoscDobraPM10.item()

```

```python
plt.style.use("fivethirtyeight")

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

values2_5 = [valDobra2_5, valSrednia2_5, valZla2_5]
values10 = [valDobra10, valSrednia10, valZla10]

suma2_5 = valDobra2_5 + valSrednia2_5 + valZla2_5
suma10 = valDobra10 + valSrednia10 + valZla10

colors = ['green', 'yellow', 'red']

labels = ["Wartość Dobra ", "Wartość Średnia", "Wartość Zła"]

ax1.set_title(f"Procentowy podział wartości powietrza wśród wszystkich pomiarów PM2.5 ({suma2_5})")
ax1.pie(values2_5,  wedgeprops={'edgecolor': 'black'}, shadow=True, startangle=45, autopct='%1.1f%%', pctdistance=1.2, colors = colors)
ax1.legend(loc=(0.9, 0.1), labels=labels)

ax2.set_title(f"Procentowy podział wartości powietrza wśród wszystkich pomiarów PM10 ({suma10})")
ax2.pie(values10,  wedgeprops={'edgecolor': 'black'}, shadow=True, startangle=45, autopct='%1.1f%%', pctdistance=1.2, colors = colors)
ax2.legend(loc=(0.9, 0.1), labels=labels)
```
